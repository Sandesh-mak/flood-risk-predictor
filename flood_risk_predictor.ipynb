```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# -----------------------------
# STEP 1: Generate Synthetic Weather Data
# -----------------------------

np.random.seed(42)
n_samples = 1000

rainfall = np.random.uniform(0, 300, n_samples)  # mm
temperature = np.random.uniform(20, 45, n_samples)  # °C
humidity = np.random.uniform(40, 100, n_samples)  # %
drainage_score = np.random.randint(1, 6, n_samples)  # 1 (poor) to 5 (good)

# Simple rule-based flood risk logic
flood_risk = []

for r, t, h, d in zip(rainfall, temperature, humidity, drainage_score):
    risk_score = (r * 0.6) + (h * 0.2) - (d * 20)

    if risk_score > 150:
        flood_risk.append("High")
    elif risk_score > 80:
        flood_risk.append("Medium")
    else:
        flood_risk.append("Low")

# Create DataFrame
data = pd.DataFrame({
    "Rainfall": rainfall,
    "Temperature": temperature,
    "Humidity": humidity,
    "DrainageScore": drainage_score,
    "FloodRisk": flood_risk
})

# -----------------------------
# STEP 2: Train Model
# -----------------------------

X = data[["Rainfall", "Temperature", "Humidity", "DrainageScore"]]
y = data["FloodRisk"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

# -----------------------------
# STEP 3: Evaluate Model
# -----------------------------

predictions = model.predict(X_test)
print("Model Evaluation:\n")
print(classification_report(y_test, predictions))

# -----------------------------
# STEP 4: User Prediction
# -----------------------------

print("\n--- Flood Risk Prediction ---")

user_rainfall = float(input("Enter rainfall (mm): "))
user_temp = float(input("Enter temperature (°C): "))
user_humidity = float(input("Enter humidity (%): "))
user_drainage = int(input("Enter drainage score (1-poor to 5-good): "))

user_data = np.array([[user_rainfall, user_temp, user_humidity, user_drainage]])
prediction = model.predict(user_data)

print("\nPredicted Flood Risk:", prediction[0])
```
